```python
import pandas as pd
```

```python
insdnames = (r'C:\Users\mcveigh\Documents\PythonPC\missing_names')
insdnames_df = pd.read_csv(insdnames,
                       sep='\t',
                       index_col=None,
                       low_memory=False,
                       usecols=[0,1,2],
                       header=None,
                       names=["accession", "organism","type"])
insdnames_df = insdnames_df[1:]
insdnames_df.head(20)
```

```python
refseqnames = (r'C:\Users\mcveigh\Documents\PythonPC\refseq_names')
refseqnames_df = pd.read_csv(refseqnames,
                       sep='\t',
                       index_col=None,
                       low_memory=False,
                       usecols=[0,1,2],
                       header=None,
                       names=["NRaccession", "organism","type"])
refseqnames_df = refseqnames_df[1:]
refseqnames_df
```

```python
refseqnames_df["binomial"] = refseqnames_df["organism"].str.split().str[:2].str.join(sep=" ")
refseqnames_df
```

### try merging on organism names

```python
combine_df=pd.merge(left=insdnames_df, right=refseqnames_df, left_on='organism', right_on='organism', how = 'outer')
#combine_df.drop_duplicates(subset=['type',], inplace=True)
combine_df.drop(columns=['type_y'], inplace=True)
combine_df
```

```python
missing_df = combine_df[combine_df['NRaccession'].isna()]
missing_df
```

## List comprehension to compare two lists and find accessions that are missing from the larger list

### Can be used for a reject list or simple sorting. Works on simple itemized lists of accession numbers

### Read file containing accessions you want removed from the larger list

```python
reffile = open("GBRefSeq_orig", "r")
refacclist = reffile.readlines()
#print(refacclist)
```

### Read file containing the full list of accessions that you want to search

```python
allfile = open("16Sacc_anylength", "r")
allacclist = allfile.readlines()
#print(allacclist)
```

```python
reject = open("16Srejectlist.txt", "r")
rejectlist = reject.readlines()
```

### Compare allacclist to refacclist and report the ones not found in refacclist

```python
missing = []
missingcount = 0
for i in range(len(allacclist)):
        if allacclist[i] not in refacclist:
            #print("missing",allacclist[i])
            missing.append(allacclist[i])
            missingcount = missingcount + 1
print(missingcount)
```

### Second search to remove any accessions found on the reject list

```python
notreject = []
notrejectnum = 0
for i in range(len(missing)):
        if missing[i] not in rejectlist:
            #print("missing",allacclist[i])
            notreject.append(missing[i])
            notrejectnum = notrejectnum + 1
print(notrejectnum)
```

```python
print(notreject)
```

### Write the list of missing accessions to a file

```python
with open(r'missingaccesssion.txt', 'w') as fp:
    for item in missing:
        # write each item on a new line
        fp.write("%s" % item)
```

### Alternate write the list after the comparision with the reject list

```python
with open(r'missingacc_notreject.txt', 'w') as fp:
    for item in notreject:
        # write each item on a new line
        fp.write("%s" % item)
```
